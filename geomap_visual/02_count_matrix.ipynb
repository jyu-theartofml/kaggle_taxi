```python
import pandas as pd
import numpy as np
import codecs, json
```

```python
location_df1=pd.read_csv('pickup_dropoff_df.csv')
location_df2=pd.read_csv('pickup_dropoff_df2.csv')
location_df3=pd.read_csv('pickup_dropoff_df3.csv')
```

```python
location_df=location_df1.append(location_df2).append(location_df3)
print(location_df.shape)
location_df.head()
```

```python
data_counts=pd.DataFrame({'count':location_df.groupby(['pickup_location', 'dropoff_location']).
                           size()}).reset_index()
```

```python
data_matrix=data_counts.pivot(index='pickup_location', columns='dropoff_location', values='count')
```

```python
data_matrix=data_matrix.fillna(value=0)
```

```python
data_matrix
```

```python
##find index names that also match column names
matching_cols = [s for s in list(data_matrix.index.values) if s in list(data_matrix.columns.values)]
```

```python
select_matrix=data_matrix.loc[matching_cols][matching_cols]
select_matrix # nxn matrix
```

```python
p=np.array(select_matrix)
b=p.tolist()
json.dump(b, codecs.open('location_matrix.json', 'w', encoding='utf-8'), separators=(',', ':'), sort_keys=True, indent=4)
```

```python
cities=pd.DataFrame(select_matrix.index)
```

```python
cities.to_csv('cities_select_matrix.csv', index=False)
```
