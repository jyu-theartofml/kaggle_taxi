```python
from pygeocoder import Geocoder
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train_data2.csv', low_memory=False)
```

```python
train.head()
```

```python
geo_data=train[3000:4000]
geo_data.head()
```

```python
pickup_point=[]
dropoff_point=[]
for row in geo_data.itertuples():
    #7th column is pick longitude, 8th is pickup latitude
    pickup_long=row[3]
    pickup_lat=row[4]
    dropoff_long=row[5]
    dropoff_lat=row[6]
    
    geolocator = Nominatim()
    pickup_query= geolocator.reverse(str(pickup_lat)+','+str(pickup_long), timeout=30)
    pickup_city=pickup_query.raw['address']
    
    geolocator = Nominatim()
    dropoff_query= geolocator.reverse(str(dropoff_lat)+','+str(dropoff_long), timeout=30)
    dropoff_city=dropoff_query.raw['address']
    
    if 'neighbourhood' in pickup_city.keys() & dropoff_city.keys():
        pickup_point.append(pickup_city['neighbourhood'])
        dropoff_point.append(dropoff_city['neighbourhood'])
```

```python
location_df=pd.DataFrame({'pickup_location':pickup_point, 'dropoff_location': dropoff_point})
location_df.shape
```

```python
location_df=location_df[['pickup_location', 'dropoff_location']]
location_df.head()
```

```python
location_df.to_csv('pickup_dropoff_df3.csv', index=False)
```
